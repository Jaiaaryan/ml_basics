```python
from matplotlib import pyplot as plt
from ipywidgets import interact
from IPython.display import clear_output
import numpy as np
```

```python
X = np.random.randn(20)
Y = .8 * X**2 + np.random.randn(20) * .2 - 2

fig = plt.scatter(X, Y)
```

```python
from ipywidgets import interact

def test_mb(m, b):
    T = np.linspace(-2, 2, 100)
    fig = plt.plot(T, m*T+b, c='r')
    fig = plt.scatter(X, m*X+b, c='r')
    fig = plt.scatter(X, Y)
    plt.grid()
    plt.show()
    #cost = np.square(m*X+b - Y)
    
interact(test_mb, m=(-5,5,.1), b=(-5,5,.1))
```

```python
# fit
m = -1
b = 1

for i in range(20):
    pred = m*X+b
    cost = np.average(np.square(pred - Y))
    dcost_dpred = 2 * (pred - Y)
    dpred_dm = X
    dpred_db = 1

    dcost_dm = np.dot(dcost_dpred, dpred_dm)
    dcost_db = np.sum(dcost_dpred * dpred_db)

    m -= .01 * dcost_dm
    b -= .01 * dcost_db
    
    clear_output(wait=True)
    T = np.linspace(-2, 2, 100)
    fig = plt.plot(T, m*T+b, c='r')
    fig = plt.scatter(X, m*X+b, c='r')
    fig = plt.scatter(X, Y)
    plt.grid()
    plt.show()
    print(m, b)
```

```python
# linear

m, b = np.polyfit(X, Y, 1)

T = np.linspace(-2, 2, 100)
fig = plt.plot(T, m*T+b, c='r')
fig = plt.scatter(X, m*X+b, c='r')
fig = plt.scatter(X, Y)

print(m, b)
```

```python
#quadratic

a, b, c = np.polyfit(X, Y, 2)

T = np.linspace(-2, 2, 100)
fig = plt.plot(T, a*T**2 + b*T + c, c='r')
fig = plt.scatter(X, a*X**2 + b*X + c, c='r')
fig = plt.scatter(X, Y)

print(m, b)
```
